# Q. 결측치 데이터(행)을 제거하고, 앞에서부터 60% 데이터만 활용해, 'f1' 컬럼 3사분위 값을 구하시오 
- 60%가 소수점일 경우 절사(예: 36.6 일때 36으로 계산)
- data: t1-data1.csv

```python
import pandas as pd
df = pd.read_csv("../input/big-data-analytics-certification/t1-data1.csv")
df
```

# Your Code

```python
df = df.dropna(axis=0)
print(df.info())

print("=========")
# 앞에서부터 60% 데이터만 활용해, 'f1' 컬럼 3사분위 값을 구하시오
df_len = len(df)
df_len_60 = int(df_len * 0.6)
df = df[:df_len_60]

q3 = df['f1'].quantile(0.75)
print("정답 : ", q3)
```

# solution

```python
df = pd.read_csv("../input/big-data-analytics-certification/t1-data1.csv")
df = df.dropna()
df = df.iloc[:int(len(df)*0.6)]
print(df['f1'].quantile(.75))
```
